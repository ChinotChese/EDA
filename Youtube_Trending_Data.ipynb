## How many trending video contain capitalized word?

If you upload videos on YouTube you may be tempted to put out some videos that capitalised titles. “It will definitely draw a viewer’s attention” you tell yourself as you hit the upload button.

Now we are going to see whether those trending videos have at least one capitalized word. In order to do that, we will add a new column to the dataset whose value is True if the video title has at least a capitalized word in it, and False otherwise.

```python
# function to check if the string contains any capitalized word
def contains_capitalized_word(s):
    for w in s.split():
        if w.isupper():
            return True
    return False

US_df["contains_capitalized"] = US_df["title"].apply(contains_capitalized_word)

value_counts = US_df["contains_capitalized"].value_counts()

fig, ax = plt.subplots(figsize=(7, 7))
plt.pie(value_counts, labels=['No', 'Yes'], colors=['#FFCC99', '#99CCCC'])
ax.set_title('Title contains at least one capitalized word')
plt.show()
```

## How long does it take for a video to become trending?

So let's see how long it takes one video to get in the top trending

```python
US_df['trending_date'] = pd.to_datetime(US_df['trending_date'], format='%y.%d.%m')
US_df['publish_time'] = pd.to_datetime(US_df['publish_time'])
US_df["publishing_date"] = pd.to_datetime(US_df["publish_time"].dt.date)

# days that each video needs to become trending
days = (US_df['trending_date'] - US_df['publishing_date']).dt.days
days
```

```python
fig, ax = plt.subplots(figsize=(14,8))
sns.boxplot(days, color='#e03531', medianprops={'color': '#FBA91A'})
plt.tight_layout()
```

Since there are a lot of videos that need even more than **4000** days to become trending, the boxplot above doesn't give us much information. These videos were published many years ago and suddenly became a hit in 2018. So we only focus on those videos which need less than **100** days in order to see the trend clearly.

```python
days = np.array(days)

count = 0
for i in range(days.shape[0]):
    if days[i] > 100:
        count += 1

print(count)
print(count/days.shape[0])
```

As you can see, number of videos take more than 100 days only account for **0.71%**


```python
new_day  = np.zeros((days.shape[0] - count))
j = 0
for i in range(days.shape[0]):
    if days[i] > 100:
        continue
    new_day[j] = days[i]
    j += 1

    
fig, ax = plt.subplots(figsize=(14,8))
sns.boxplot(new_day, color='#e03531', medianprops={'color': '#FBA91A'})
plt.tight_layout()
```

```python
pd.DataFrame(new_day).describe()
```

So on average, it takes approximately **6** to **7** days for a video to get in the top trending

* **`68%`** of trending videos have less than **`40,000`** likes, and **`84%`** have less than **`100,000`** likes.
* **`72.2%`** of trending videos have less than **`5,000`** comments, and **`83.8%`** have less than **`10,000`** comments.
* Some videos may appear on the trending videos list on more than one day. Our dataset contains **`40494`** entries but not for **`40494`** unique videos but for **`6351`** unique videos.
* Trending videos have title length between **`30`** and **`50`** characters approximately.
* The words **'Official'**, **'Video'**, **'Trailer'**, **'How'**, and **'2018'** were common also in trending video titles.
* There is a strong positive correlation between the number of views and the number of likes of trending videos: As one of them increases, the other increases, and vice versa.
* There is a strong positive correlation also between the number of **`likes`** and the number of **`comments`**, and a slightly weaker one between the number of **`dislikes`** and the number of **`comments`**.
* The category that has the largest number of trending videos is **'Entertainment'** with **`9,964`** videos, followed by 'Music' category with **`6,472`** videos, followed by 'Howto & Style' category with **`4,146`** videos.
* On the opposite side, the category that has the smallest number of trending videos is **'Shows'** with **`57`** videos, followed by **'Nonprofits & Activisim'** with **`57`** videos, followed by **'Autos & Vehicles'** with **`384`** videos.
* About the tags used in trending videos: The average number of tags a trending video has is **`21`** tags. Only **`3.5%`** of trending videos have no tags. The minimum number of tags a trending video has is **`0`**. The maximum number of tags a trending video has is **`73`**.
*On average, the length of the tags of a trending video is **`256`** characters
